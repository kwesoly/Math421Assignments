## Bringing in the Dataset

```r
adult = read.csv("C:/Users/student/Documents/Adult.csv", header=FALSE, na.strings = c(" ?", "NA", "."))

names(adult)

names(adult) = c("Age", "Working Class", "Final Weight", "Education", "Ed", "Marital Status", "Occupation", "Relationship", "Race", "Sex", "Gains", "Losses", "AvgHrsWeek", "Country", "target")

for(i in 1:ncol(adult)) 
    {
    adult[adult ==" ?"]= NA  
    missval <- sum(is.na(adult[i]))
    if(missval > 0)
      {
      if(is.numeric(adult[[i]]) == TRUE) 
        {
         adult[[i]][is.na(adult[[i]])] <- mean(adult[[i]], na.rm = TRUE)
        }
        else 
        {
        levels=unique(adult[,i])
        adult[,i][is.na(adult[,i])]=levels[which.max(tabulate(match(adult[,i], x = levels)))]
      }
    }
}
```
