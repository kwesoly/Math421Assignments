```r
knitr::opts_chunk$set(echo = TRUE)
library(ggplot2)
library(readxl)
train <- read_excel("C:/Users/student/Desktop/train.xls")
train$Survived = factor(train$Survived)
train$Pclass = factor(train$Pclass)
train$Sex = factor(train$Sex)
train$Embarked = factor(train$Embarked)
```

## R Markdown

### Survived - Graph 1

```r
ggplot(train, aes(x = Survived)) + theme_bw() + geom_bar() + labs(y = "Passenger Count", title = "Survival Rates")
prop.table(table(train$Survived))
```

#### The graph above simply shows the difference between the number of passengers who survived on the boat and the number of passengers that did not. 0 Represents the passengers who died and 1 represents the survivers. The percentages are also shown above, showing about 62% died while about 38% survived.

### Survival Rates by Sex - Graph 2 

```r
ggplot(train, aes(x = Sex, fill = Survived)) + theme_bw() + geom_bar() + labs(y = "Passenger Count", title = "Survival Rates by Sex")
```

#### This graph explains the survival rate based on gender. The orange color represents those who did not survive and the teal color represents those who did. This shows that more females survived than did not, while more males did not survive versus survived. This proves the theory that the women and children were saved first.

### Survival Rate with Pclass - Graph 3

```r
ggplot(train, aes(x = Pclass, fill = Survived)) + theme_bw() + geom_bar() + labs(y = "Passenger Count", title = "Survival Rates by Pclass")
```

#### This shows the survival rate on the titanic by class of ticket.

### Pclass & Sex - Graph 4

```r
ggplot(train, aes(x = Sex, fill = Survived)) + theme_bw() + facet_wrap(~ Pclass) + geom_bar() + labs(y = "Passenger Count", title = "Survival Rates by Pclass and Sex")
```

#### This graph not only shows the survival rate on the titanic by class of ticket but in relation to Sex, as well. Each class is represented in a panel. 

### Age Distribution - Graph 5

```r
ggplot(train, aes(x = Age)) + theme_bw() + geom_histogram(binwidth = 5) + labs(y = "Passenger Count", x = "Age (binwidth = 5)", title = "Age Distribution")
```

#### This graph shows the distribution of passenger ages. I used a width of 5 years because that increment because it worked out better than 10 or 1 year increments. Because there is missing data in the Age column I received a warning message saying it had removed 177 rows. 

### Survival Rates by Age - Graph 6

```r
ggplot(train, aes(x = Age, fill = Survived)) + theme_bw() + geom_histogram(binwidth = 5) + labs(y = "Passenger Count", x = "Age (binwidth = 5)", title = "Survival Rates by Age")
```

#### Once again we are missing 177 values. This graph is providing more information than the last graph because it's showing the proportion between the ones who died between the ones who survived. In the proportion of children, more than half survived but at the other end with the older people, the rate of surviving is pretty low. However, there was one outlier where there was one 80 year old person who survived. 

### Box and Whisker Survival Rates by Age - Graph 7

```r
ggplot(train, aes(x = Survived, y = Age)) + theme_bw() + geom_boxplot() + labs(y = "Age", x = "Survived", title = "Survival Rates by Age")
```

#### This is another way to show survival rates based on age. However, it is helpful to compare with the previous graph and does not tell as much of a story by itself.
